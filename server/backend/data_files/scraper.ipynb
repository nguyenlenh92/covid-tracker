```python
# Import pandas and json
import pandas as pd

#get the file from github and read it
!curl https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-counties.csv > ~/counties.csv
df = pd.read_csv('~/counties.csv')

#remove US territories
territories = ['Puerto Rico','Virgin Islands','Guam','Northern Mariana Islands','American Samoa']
for x in territories:
    df.drop(df.index[df["state"] == x], inplace=True)

#remove columns that we will not use
df.drop(df.index[df["county"] == 'Unknown'], inplace=True)
df.drop(df.index[df["county"] == ' '], inplace=True)
df.drop(df.index[df["fips"].isnull()],inplace=True)
df.drop(columns=['state','county'],inplace=True)
df.drop(columns='probable_cases', inplace=True)
df.drop(columns='confirmed_cases', inplace=True)
df.drop(columns='probable_deaths', inplace=True)
df.drop(columns='confirmed_deaths', inplace=True)

#case columns as int
df['fips'] = df['fips'].astype(int)
df['deaths'] = df['deaths'].astype(int)
df['cases'] = df['cases'].astype(int)


#remove the case columns from death
deathDf = df.copy(deep=True)

deathDf.drop(columns='cases', inplace=True)
df.drop(columns='deaths', inplace=True)
df
```

```python
#create death csv
deathDf.to_csv("deaths.csv",index=False)
```

```python
#create cases csv
df.to_csv("cases.csv",index=False) 
```

```python
!curl https://raw.githubusercontent.com/do71730/data/main/vaccination.csv> ~/countiesVac.csv
df2 = pd.read_csv('~/countiesVac.csv')

#remove US territories
territories2 = ['AS','GU','MP','PR','VI']
for x in territories2:
    df2.drop(df2.index[df2["Recip_State"] == x], inplace=True)

#remove unknown fips
df2.drop(df2.index[df2["FIPS"] == "UNK" ],inplace=True)
df2['FIPS'] = df2['FIPS'].astype(int)
df2['Series_Complete_Yes'] = df2['Series_Complete_Yes'].astype(int)

df2.drop(df2.loc[:,'Series_Complete_12Plus' :  ].columns,axis=1,inplace=True)
df2.drop(columns=['MMWR_week','Series_Complete_Pop_Pct','Recip_State','Recip_County'], inplace=True)
df2['Date'] = df2['Date'].str.replace('/','-')
df2.rename(columns={'FIPS' : 'fips','Series_Complete_Yes' : 'vaccine'},inplace=True)
df2
```

```python
#create vaccine csv
df2.to_csv("vaccine.csv",index=False)
```
